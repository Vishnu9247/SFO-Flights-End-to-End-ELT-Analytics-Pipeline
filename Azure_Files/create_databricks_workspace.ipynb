```python
from azure.identity import DefaultAzureCredential
from azure.mgmt.resource import ResourceManagementClient
from azure.mgmt.databricks.models import Workspace, Sku
from azure.mgmt.databricks import AzureDatabricksManagementClient
from dotenv import load_dotenv
import os
```

```python
credential = DefaultAzureCredential()
load_dotenv()
subscription_id = os.environ["AZURE_SUBSCRIPTION_ID"]
databricks_client = AzureDatabricksManagementClient(credential, subscription_id)
RESOURCE_GROUP_NAME = os.environ["AZURE_RESOURCE_GROUP_NAME"]
DATABRICKS_NAME = os.environ["DATABRICKS_NAME"]
LOCATION = os.environ["LOCATION"]
```

```python
Workspace_params = Workspace(
    location=LOCATION,
    sku=Sku(name="premium"),
    managed_resource_group_id=f"/subscriptions/{subscription_id}/resourceGroups/{RESOURCE_GROUP_NAME}-databricks-rg"
)
```

```python
workspace = databricks_client.workspaces.begin_create_or_update(
    RESOURCE_GROUP_NAME,
    DATABRICKS_NAME,
    Workspace_params
).result()

print(f"Databricks workspace {DATABRICKS_NAME} created in resource group {RESOURCE_GROUP_NAME}.")
```
